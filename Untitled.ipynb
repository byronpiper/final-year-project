```python
import numpy as np
import matplotlib.pyplot as plt

def generate_naca_airfoil(digit1, digit2, digit3, digit4, num_points=100):
    # Generate evenly spaced points from 0 to 1
    x = np.linspace(0, 1, num_points)
    
    # Calculate thickness distribution
    thickness = (10 * digit1 * (0.1 * x)**0.5 -
                 20 * digit2 * (0.1 * x)**2 +
                 10 * digit3 * (0.1 * x)**3 +
                 (0.1 * digit4) * (0.1 * x)**4)
    
    # Calculate camber line for symmetric airfoil
    camber = np.zeros_like(x)
    
    # Combine thickness and camber to get upper and lower surfaces
    xu = x - thickness * np.sin(0)
    xl = x + thickness * np.sin(0)
    yu = camber + thickness * np.cos(0)
    yl = camber - thickness * np.cos(0)
    
    return xu, xl, yu, yl

def plot_airfoil(xu, xl, yu, yl):
    plt.figure(figsize=(100, 6))  # Adjust the width and height here
    plt.plot(xu, yu, 'b-', label='Upper Surface')
    plt.plot(xl, yl, 'r-', label='Lower Surface')
    plt.gca().set_aspect('equal')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('NACA Airfoil')
    plt.legend()
    plt.grid(True)
    plt.show()

def generate_random_naca_airfoil():
    digit1 = np.random.randint(0, 10)
    digit2 = np.random.randint(0, 10)
    digit3 = np.random.randint(0, 10)
    digit4 = np.random.randint(0, 10)
    return generate_naca_airfoil(digit1, digit2, digit3, digit4)

if __name__ == "__main__":
    xu, xl, yu, yl = generate_random_naca_airfoil()
    plot_airfoil(xu, xl, yu, yl)
```
